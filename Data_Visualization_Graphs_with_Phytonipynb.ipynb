# Understanding the data to start processing and analysing it
Analysis of Brazil's immigration trends over the years

```python
df.set_index('Country', inplace=True) # Adjusting the index for countries will help to filter out Brazil later on
df
```

```python
years = list(map(str, range(1980, 2014))) # Creating a list to make it easier to group the years in the years columns
years
```

```python
brazil = df.loc['Brazil', years] # Filter by country and year, according to the previously created variable
brazil
```

```python
brazil_dict = {'year': brazil.index.tolist(), 'immigrants': brazil.values.tolist()}
data_brazil = pd.DataFrame(brazil_dict)  # Create a data frame applying the years and immigrants columns to understand the behavior of the data.
data_brazil
```

# Matplotlib Library

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6)) # Graphic size definition
plt.plot(data_brazil['year'], data_brazil['immigrants'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) # Adjust to look at the data on the X-axis every 5 years
plt.yticks([500, 1000, 1500, 2000, 2500, 3000]) # Adjust to look at the data on the Y-axis every 500 values
plt.xlabel('Year') # y-axis label
plt.ylabel('Number of Immigrants') # x-axis label
plt.title('Immigration from Brazil to Canada over the years') # Definition of Title
plt.show()
```

#  Extra exercise - Comparing two different countries
Line graph analysis to compare the number of immigrants from Brazil and Argentina to Canada

```python
df_comparation = df.loc[['Brazil', 'Argentina'], years] # Filter by country and year, according to the previously created variable of years
df_comparation
```

```python
df_comparation = df_comparation.T  # Transpose rows into columns
df_comparation
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_comparation['Brazil'],label = 'Brazil')
plt.plot(df_comparation['Argentina'],label ='Argentina')
plt.title('Immigration from Brazil and Argentina to Canada')
plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

# More customization
Back to Brazil analyses

```python
fig, ax = plt.subplots(figsize=(8, 4)) # Graphic size definition as a image
ax.plot(data_brazil['year'], data_brazil['immigrants'], lw=3, marker='o', color='g')
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # frequency every 5 years
ax.set_xlabel('Year', fontsize=14) # x-axis label
ax.set_ylabel('Number of Immigrants', fontsize=14) # y-axis label
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.set_title('Immigration from Brazil to Canada\n 1980 to 2013', fontsize=18) # insert a title
plt.grid(linestyle='--')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('Immigration_Brazil_Canada.png', transparent=False, dpi=300, bbox_inches='tight') # Save the image

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].plot(data_brazil['year'], data_brazil['immigrants'], color='g')
axs[0].set_title('Immigration from Brazil to Canada\n 1980 to 2013')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Number of Immigrants')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid(True)

axs[1].boxplot(data_brazil['immigrants'])
axs[1].set_title('Boxplot of Immigration from Brazil to Canada\n 1980 to 2013')
axs[1].set_xlabel('Brazil')
axs[1].set_ylabel('Number of Immigrants')
axs[1].grid(True)

plt.show()
```

# Subplots Analyses
Immigration trends from South America in Brazil, Colombia, Argentina and Peru

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6)) # Size adjustment
fig.subplots_adjust(hspace=0.5, wspace=0.3) # Horizontal and vertical spacing adjustment

fig.suptitle('Immigration from the biggest countries from Latin America to Canada - 1890 to 2013')

axs[0,0].plot(df.loc['Brazil', years])
axs[0,0].set_title('Brazil')
axs[0,0].set_ylabel('Number of Immigrants')

axs[0,1].plot(df.loc['Colombia', years])
axs[0,1].set_title('Colombia')

axs[1,0].plot(df.loc['Argentina', years])
axs[1,0].set_title('Argentina')
axs[1,0].set_xlabel('Year')
axs[1,0].set_ylabel('Number of Immigrants')

axs[1,1].plot(df.loc['Peru', years])
axs[1,1].set_title('Peru')
axs[1,1].set_xlabel('Year')

for ax in axs.flat:
     ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # Frequency every 5 years

ymin = 0
ymax = 7000
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
south_american = df.query('Region == "South America"')
south_american
```

```python
colors = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
```

```python
fig, axs = plt.subplots(figsize=(12,5))
axs.bar(south_american.index, south_american['Total'], color=colors)

axs.set_title('Immigration from South America to Canada - Total', fontsize=18)
axs.set_xlabel('Year', fontsize=14)
axs.set_ylabel('Number of Immigrants', fontsize=14)
axs.xaxis.set_tick_params(labelsize=12)
axs.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
south_american.sorted = south_american.sort_values('Total', ascending=True)
south_american.sorted
```

```python
colors = []
for county in south_american.sorted.index:
    if county == 'Brazil':
      colors.append('green')
    else:
      colors.append('silver')


fig, axs = plt.subplots(figsize=(12,5))
bars = axs.barh(south_american.sorted.index, south_american.sorted['Total'], color=colors)

# Configurações do gráfico
axs.set_title('Brazil was the 4th country with the most immigrants\nto Canada from 1980 to 2013', fontsize=18, loc='left')
axs.set_xlabel('Number of Immigrants', fontsize=14)
axs.set_ylabel('', fontsize=14)
axs.xaxis.set_tick_params(labelsize=12)
axs.yaxis.set_tick_params(labelsize=12)


for i, v in enumerate(south_american.sorted['Total']):
    axs.text(v + 20, i, str(v), color='black', fontsize=8, ha='left', va='center', fontweight='bold')

axs.set_frame_on(False)
axs.get_xaxis().set_visible(False)
axs.tick_params(axis='both', which='both', length=0)

fig.savefig('Immigration_South_America.png', transparent=False, dpi=300, bbox_inches='tight') # Save the image

plt.show()
```

# Customizing colors

```python
def generate_palette_graphic(palette):
  fig, ax = plt.subplots(figsize=(8, 4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
  ax.set_title('Top 10 Countries with the Highest Immigration\n to Canada from 1980 to 2013', fontsize=16, loc='left')
  ax.set_xlabel('Number of Immigrants', fontsize=14)
  ax.set_ylabel('', fontsize=14)
  plt.show()
```

```python
import plotly.express as px
```

```python
fig = px.line(data_brazil, x='year', y='immigrants',
              title='Immigration from Brazil to Canada from 1980 to 2013')
fig.update_traces(line_color='green', line_width=4)
fig.update_layout(xaxis_title='Year', yaxis_title='Number of Immigrants',
                  width=1000, height=500,
                  xaxis={'tickangle': -45},
                  font_family='Arial',
                  font_size=14,
                  font_color="grey",
                  title_font_color='black',
                  title_font_size=20,
                  title_font_family='Arial'
                  )
fig.show()
```

# Conclusion

From 1980 to 2013, Brazil experienced distinct immigration patterns. Throughout the 1980s and 1990s, there was only one notable peak in 1992, with 1,066 immigrants per year. However, a significant resurgence occurred in 2005-2006, with immigration rates exceeding 1,000 immigrants annually and continuing to climb until reaching a high point in 2010 with 2,598 immigrants. In 2011, there was a notable drop to 1,508 immigrants, though levels stabilized in subsequent years, not falling below 1,000 per year again.

This trend invites further analysis into the causes behind the immigration peak from the early 2000s through 2010, examining possible internal and international political factors that may have influenced this marked increase.

Other major countries demonstrated varied patterns:

* Argentina followed a pattern similar to Brazil, though its immigration peak occurred in 2003 with 1,745 immigrants, followed by a decline that continued steadily until the end of the data period.
* Chile exhibited a different pattern, with notable peaks in 1987 (1,454 immigrants) and again in 1991 (1,749 immigrants), its highest point. After a steep decline from 1992 to 1994, immigration levels stabilized around 250-300 immigrants per year.
* Colombia showed the highest immigration volume, surpassing 6,000 immigrants in both 2005 and 2006, with an all-time peak of 6,535 immigrants. Colombian immigration began rising around 1997-1998, peaking in the mid-2000s before a notable decline in 2011, which continued gradually through the end of the period, yet still remained significantly higher than that of any other country analyzed.
