```python
import pandas as pd
```

```python
df = pd.read_csv('/content/canadian_immegration_data.csv')
df
```

```python
df.info() # Check the data types to see if you need to handle null data or data format, for example
```

EN: Analysis of Brazil's immigration trends over the years //  PT: Analise de tendência de imigração do Brasil ao longo dos anos

```python
df.set_index('Country', inplace=True) # Adjusting the index for countries will help to filter out Brazil later on
df
```

```python
years = list(map(str, range(1980, 2014))) # Creating a list to make it easier to group the years in the years columns
years
```

```python
brazil = df.loc['Brazil', years] # Filter by country and year, according to the previously created variable
brazil
```

```python
brazil_dict = {'year': brazil.index.tolist(), 'immigrants': brazil.values.tolist()}
data_brazil = pd.DataFrame(brazil_dict)  # Create a data frame applying the years and immigrants columns to understand the behavior of the data.
data_brazil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6)) # Graphic size definition
plt.plot(data_brazil['year'], data_brazil['immigrants'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) # Adjust to look at the data on the X-axis every 5 years
plt.yticks([500, 1000, 1500, 2000, 2500, 3000]) # Adjust to look at the data on the Y-axis every 500 values
plt.xlabel('Year') # y-axis label
plt.ylabel('Number of Immigrants') # x-axis label
plt.title('Immigration from Brazil to Canada over the years') # Definition of Title
plt.show()
```

Line graph analysis to compare the number of immigrants from Brazil and Argentina to Canada

```python
df_comparation = df.loc[['Brazil', 'Argentina'], years] # Filter by country and year, according to the previously created variable of years
df_comparation
```

```python
df_comparation = df_comparation.T  # Transpose rows into columns
df_comparation
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_comparation['Brazil'],label = 'Brazil')
plt.plot(df_comparation['Argentina'],label ='Argentina')
plt.title('Immigration from Brazil and Argentina to Canada')
plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```
