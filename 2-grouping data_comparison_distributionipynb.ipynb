# Question 1: Whats are the total sales per year? Witch year performed best in our databse?

```python
df_sales_year = sales.copy()
df_sales_year = df_sales_year[["data_pedido", "vendas"]] # filter columns of interest

df_sales_year["year"] = df_sales_year["data_pedido"].dt.year # create a new year column
df_sales_year.drop(labels = "data_pedido", axis = 1, inplace=True) # drop just the column data_pedido

df_sales_year = df_sales_year.groupby("year").sum() # group by year and sum

df_sales_year
```

```python
def sales_visualization(color: list=[BLUE2]):

    # Import librarys of visualization
  import matplotlib.pyplot as plt
  import seaborn as sns

  fig, ax = plt.subplots(figsize=(10,4)) # define graph and theme of visualization
  sns.set_theme(style="white") # define standart style

  ax = sns.barplot(data = df_sales_year, y='vendas', x = df_sales_year.index, palette = color)

  ax.set_title("Department stores' sales \n 2016 to 2049", loc="left", fontsize = 18, color = GREY1) # define title
  ax.set_xlabel(" ") # define x label
  ax.set_ylabel(" ") # define y label
  ax.xaxis.set_tick_params(labelsize=14, labelcolor=GREY2) # define x axis tick size and color
  sns.despine(left=True, bottom = True) # remove spines

  ax.set_yticklabels([]) # remove y tick label
  for i, valor in enumerate(df_sales_year["vendas"]):
    qtd = f'R$ {valor:,.0f}'.replace(",",".")
    offset = 1e5
    ax.text(i, valor + offset, qtd, color = GREY2, fontsize = 12, ha = "center", va = "center") # create data labels

  return ax

ax = sales_visualization()
```

```python
colors = []
for year in df_sales_year.index:
  if df_sales_year.loc[year, "vendas"] == df_sales_year["vendas"].max():
    colors.append(BLUE2)
  else:
    colors.append(BLUE4)

ax = sales_visualization(colors) #define focus in the higgest value (2019)

ax.text(3.5, 1.5e6,
         'In $\\bf{2019}$, store\n'
         'sales increased\n'
         'arround $\\bf{22,3}$%\n'
         'compared to 2018.',
         fontsize=14, linespacing=1.45, color=BLUE2) # define comment

fig = ax.get_figure()
```

# Question 2: What are the top 7 most profitable products in our catalogue during the period shown?


```python
top_profit = sales.copy()
top_profit = top_profit[["tipo_produto", "lucro"]] # filter columns of interest

top_profit = top_profit.groupby("tipo_produto").sum() #grouping values
top_profit = top_profit.sort_values(by="lucro", ascending=False) # put in ascending format
top_profit = top_profit.head(7)

top_7 = top_profit[:7] # preparing the table that will generate the graph
top_7
```

```python
def top_7_visualization():

  import matplotlib.pyplot as plt
  import seaborn as sns

  # Área do gráfico e tema da visualização
  fig, ax = plt.subplots(figsize=(10,4))
  fig.subplots_adjust(right=0.7)
  sns.set_theme(style="white")

  colors = [BLUE2, BLUE2, BLUE2, GREEN1, GREEN1, GREY3, GREEN1]

  ax = sns.barplot(data = top_7, x = "lucro", y = top_7.index, palette = colors)

  ax.set_title('Top 7 most profitable products in the catalogue \n(2016-2019)\n', fontsize=18, color=GREY1, loc='left')
  ax.set_xlabel('')
  ax.set_ylabel('')
  ax.set_xticklabels([])
  ax.yaxis.set_tick_params(labelsize=10, labelcolor = GREY2 )
  sns.despine(left = True, bottom = True)

  for i, valor in enumerate(top_7['lucro']):
      qtd = f'R$ {valor:,.0f}'.replace(',','.')
      offset = 1e3  # offset de 1.000
      ax.text(valor - offset, i, qtd, color= GREY5, fontsize=10, fontweight='bold', ha='right', va='center')

  return ax

ax = top_7_visualization()
```

Conclusion:

There were 3 departments that made the top 7 most profitable, with Automotive items leading the way (blue), followed by Gardening and Landscaping (green) and just one product from General Items (grey) which is "Ferramentas"

# Question 3

- If we separate the total sales per year by region? What can we see in our data?

```python
df_sales_reg = sales.copy()
df_sales_reg = df_sales_reg[['data_pedido', 'regiao', 'vendas']]

df_sales_reg['year'] = df_sales_reg['data_pedido'].dt.year
df_sales_reg.drop(labels='data_pedido', axis=1, inplace=True)


df_sales_reg
```

```python
sales_reg = pd.crosstab(index = df_sales_reg['year'], columns = df_sales_reg['regiao'], values = df_sales_reg['vendas'], aggfunc='sum') # create a cross table
sales_reg
```

```python
sales_reg = sales_reg[["Sudeste", "Nordeste", "Centro-Oeste", "Norte", "Sul"]]
sales_reg
```

```python
def graph_sales_reg():
  colors = [BLUE2, BLUE3, BLUE4, GREY3, GREY4]

  ax = sales_reg.plot(kind="bar", stacked=True, figsize=(10,6), color = colors)

  ax.set_title('Department stores sales by region\nde 2016 a 2019', loc='left', fontsize=22, color = GREY1)
  ax.set_xlabel('')
  ax.set_ylabel('')

  ax.set_yticklabels([])
  ax.xaxis.set_tick_params(labelsize=14, labelcolor = GREY2, rotation = 0)
  ax.set_frame_on(False)

  ax.tick_params(axis="both", which="both", length=0)
  ax.text(3.35, 1.1e6, '87%', fontweight='bold', fontsize=20, color=BLUE1)
  ax.text(3.35, 5.5e5,
        'of sales in shops\n'
        'are concentrated in the regions\n'
        '$\\bf{South East}$, $\\bf{North-East}$ e\n'
        '$\\bf{Centre-West}$ in 2019.',
        fontsize=14,
        linespacing=1.45,
        color=BLUE1)

  return ax

ax = graph_sales_reg()
```

```python
ax = graph_sales_reg()

ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., # put the subtitle to the right
          reverse = True, # reverse order from largest to smallest > true | opposite > false
          title = 'Regions', # applying a title
          labels = ['Southeast', 'Northeast', 'Centre-West', 'North', 'South'] # rename labels
          ) # put the subtitle to the right

# inclusion of values on labels
for container in ax.containers:
  labels = [f'R${valor.get_height()/1000:,.0f}k '.replace(",",".") for valor in container]
  ax.bar_label(container, label_type="center", labels = labels, size = 10, color = GREY5, fontweight="bold")

  ax.axvline(x = 3.3, ymin=0, ymax=0.83, color=BLUE1, linewidth=1.2)


fig = ax.get_figure()
```

# Question 4:

- What is the shipping method most used by the shop's customers? Is it proportional for B2B and B2C?

```python
df_send = sales .copy()
df_send = df_send[["modo_envio", "segmento_cliente"]]

df_send_seg = pd.crosstab(index = df_send.modo_envio, columns=df_send.segmento_cliente).sort_values("B2B")
df_send_seg
```

```python
# normalise: add up the values and divide by their weights

df_send_normalise = pd.crosstab(index = df_send.modo_envio, columns=df_send.segmento_cliente, normalize = "index")

df_send_normalise
```

```python
# Função para escrita de texto utilizando diferentes cores. Não existe uma
# função nativa para isso. (Adaptado de https://github.com/empathy87/storytelling-with-data)
from matplotlib import transforms

def text_color(x, y, texto, colors, esp=20, ax=None, **kw):
    cores = list(reversed(colors))
    t = ax.transData
    canvas = ax.figure.canvas

    for i, linha in enumerate(reversed(texto.split('\n'))):
        frases = linha.split('||')
        for s, cor in zip(frases, cores[i]):
            text = ax.text(x, y, s, color=cor, transform=t, **kw)
            text.draw(canvas.get_renderer())
            ex = text.get_window_extent()
            t = transforms.offset_copy(text._transform, x=ex.width,
                                       units='dots')

        t = transforms.offset_copy(ax.transData, x=0, y=(i + 1) * esp, units='dots')

```

```python
## Configuring the graph with parameters that enhance data visualisation

import matplotlib.pyplot as plt

# Chart area and visualisation theme
fig, axs = plt.subplots(1, 2, figsize=(10,6), sharey = True)

# Defining the colours of the graph
colors = [BLUE2, BLUE5]

# GRAPH 1 - Stacked bars
axs[0].barh(df_send_seg.index, df_send_seg["B2B"], color = colors[0])
axs[0].barh(df_send_seg.index, df_send_seg["B2C"], left = df_send_seg["B2B"], color = colors[1])
axs[0].set_title('Absolute values\n', color = GREY2)
axs[0].set_xlabel('')
axs[0].set_ylabel('')
axs[0].set_xticklabels([])
axs[0].yaxis.set_tick_params(labelsize=14, color = GREY2)
axs[0].set_frame_on(False)

# remove all ticks from the x and y axis
axs[0].tick_params(axis='both', which='both', length=0)

text_color(0, 3.5, '$\\bf{B2B}$ | || $\\bf{B2C}$', colors = [colors], ax=axs[0], fontsize=12) # subtitle

#######################################

# GRAPH 2 - Bars stacked 100 %
axs[1].barh(df_send_normalise.index, df_send_normalise["B2B"], color = colors[0])
axs[1].barh(df_send_normalise.index, df_send_normalise["B2C"], left = df_send_normalise["B2B"], color = colors[1])
axs[1].set_title('Percentage\n', color = GREY2)
axs[1].set_xlabel('100%', color = GREY2, size = 14, loc='right')
axs[1].set_ylabel('')
axs[1].set_xticklabels([])
axs[1].set_frame_on(False)

# remove all ticks from the x and y axis
axs[1].tick_params(axis='both', which='both', length=0)

text_color(0, 3.5, '$\\bf{B2B}$ | || $\\bf{B2C}$', colors = [colors], ax=axs[1], fontsize=12) # subtitle

#######################################

# Graphics title
fig.suptitle("Most used shipping method by B2B/B2C customers (2016-2019)\n",
             fontsize=18, color=GREY1, ha = 'right', x = 0.95, y = 1.05)

# Bar values
for container in axs[0].containers:
    labels = [str(valor.get_width()) if valor.get_width() > 300 else '' for valor in container]
    axs[0].bar_label(container, label_type='center', labels = labels, size = 12, color = GREY5, fontweight='bold')

for container in axs[1].containers:
    labels = [f'{valor.get_width()*100:.1f}%' for valor in container]
    axs[1].bar_label(container, label_type='center', labels = labels, size = 12, color = GREY5, fontweight='bold')

fig.show()
```
