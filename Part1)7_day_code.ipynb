DAY 3 - EXPLORATORY DATA ANALYSIS WITH A FOCUS ON DATETIME AND TIMELINE GRAPHS

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
full_data['id_emprestimo'].value_counts()
```

```python
full_data.loc[full_data['id_emprestimo']==2772079]
```

```python
lending = len(full_data['id_emprestimo'].unique())
lending
```

```python
copies = len(full_data['id_exemplar'])
copies
```

```python
lending_date = pd.DataFrame(full_data['data_emprestimo'].value_counts()).reset_index()
lending_date.columns = ['Data', 'Quantidade']
lending_date
```

```python
lending_date['Data'] = pd.to_datetime(lending_date['Data'])
lending_date
```

```python
per_year = lending_date.groupby(lending_date['Data'].dt.year)['Quantidade'].sum().reset_index()
per_year.columns = ['Ano', 'Quantidade']
per_year
```

```python
sns.set_theme(context='notebook',
              style='darkgrid',
              palette='deep',
              font_scale=1.3,
              rc={"figure.figsize":(10,4)})
```

```python
ax = sns.lineplot(data=per_year,x='Ano',y='Quantidade', marker='o')
ax.set_xlabel('Ano', fontsize=12)
ax.set_ylabel('Qtde', fontsize=12)
ax.tick_params(axis='x', rotation=0)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

ax.set_title('Number of copies borrowed per year',size=10,loc='left',weight='bold',)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

ax=ax
```

**Q&A:**

* Is the number of loans increasing or decreasing over the last few years?

A: Between 2010 and 2017, lending remained constant at between 150k and 200k loans per year. However, in 2019 there was a sharp drop, bringing 2020 to less than 50k loans per year, probably due to the marked impact of the COVID-19 pandemic.

```python
per_month = lending_date.groupby(lending_date['Data'].dt.month)['Quantidade'].sum()
dictionary_months = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
per_month = per_month.rename(dictionary_months)
per_month = per_month.reset_index()
per_month.columns = ['mês','quantidade']
per_month
```

```python
sns.set_theme(context='notebook',
              style='darkgrid',
              palette='deep',
              font_scale=1.3,
              rc={"figure.figsize":(10,5)})
```

```python
ax = sns.lineplot(data=per_month,x='mês',y='quantidade', lw=2, marker='o')
ax.set_xlabel('Month', fontsize=10)
ax.set_ylabel('Vol', fontsize=10)
ax.tick_params(axis='x',rotation=0)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

ax.set_title("Number of copies borrowed from SISBI per month"+"\n",size=10,loc='left',weight='bold')
ax.text(s='Range from 2010 to 2020',x=-0.5,y=265000,fontsize=10, ha='left',color='gray')
```

**Q&A**

* What is your suggestion for managing staff holiday schedules according to the months of lowest demand?

A: On a monthly basis the variation in loans seems to vary with the university's academic year periods, taking into account from 2010 to 2020, the peak months for loans are February, March, April, August and September. While the lowest months for loans, and where there is a good opportunity to manage holidays for library staff, are January, December and July (all holiday months), for this reason June also has a fairly sharp drop because part of the month may be affected by holidays according to the university calendar.

```python
lending_date.head()
```

```python
per_hour = lending_date.groupby(by=lending_date['Data'].dt.hour)['Quantidade'].sum()
per_hour.index.name = 'hora'
per_hour = per_hour.reset_index()
per_hour.columns = ['hora','quantidade']
per_hour
```

```python
per_hour = per_hour.sort_values(ascending=True,by='quantidade')
```

```python
ax = sns.barplot(data=per_hour,y='quantidade',x='hora',
                 palette='Blues',hue='quantidade',dodge=False)
plt.legend([],[], frameon=False)

ax.set_xlabel('Time', fontsize=10)
ax.set_ylabel('Vol', fontsize=10)
ax.tick_params(axis='x',rotation=0)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
ax.set_title("Number of copies borrowed from SISBI by time slot"+"\n",size=10,loc='left',weight='bold')
ax.text(s='Range from 2010 to 2020',x=-0.5,y=225000,fontsize=10, ha='left',color='gray')

ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

ax=ax
```

**Q&A**

* What is your suggestion for scheduling non-user service activities for specific periods of lower demand?

A: The times with the highest demand for loans are between 10am and 6pm. The window of opportunity to have fewer people on duty or to dedicate library staff time to operational activities would be between 7am and 9am or from 7pm to 9pm, and there is also a small opportunity at 1pm, but this is usually a time allocated for staff lunches and breaks anyway.
