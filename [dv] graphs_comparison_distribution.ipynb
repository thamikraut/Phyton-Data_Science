```python
# Defining the colour palette
BLUE1, BLUE2, BLUE3, BLUE4, BLUE5 = '#03045e', '#0077b6', "#00b4d8", '#90e0ef', '#CDDBF3'
GREY1, GREY2, GREY3, GREY4, GREY5 = '#212529', '#495057', '#adb5bd', '#dee2e6', '#f8f9fa'
RED1, ORANGE1, YELLOW1, GREEN1, GREEN2 = '#e76f51', '#f4a261',	'#e9c46a', '#4c956c', '#2a9d8f'
```

# Question 1: Whats are the total sales per year? Witch year performed best in our databse?

```python
df_sales_year = sales.copy()
df_sales_year = df_sales_year[["data_pedido", "vendas"]] # filter columns of interest

df_sales_year["year"] = df_sales_year["data_pedido"].dt.year # create a new year column
df_sales_year.drop(labels = "data_pedido", axis = 1, inplace=True) # drop just the column data_pedido

df_sales_year = df_sales_year.groupby("year").sum() # group by year and sum

df_sales_year
```

```python
def sales_visualization(color: list=[BLUE2]):

    # Import librarys of visualization
  import matplotlib.pyplot as plt
  import seaborn as sns

  fig, ax = plt.subplots(figsize=(10,4)) # define graph and theme of visualization
  sns.set_theme(style="white") # define standart style

  ax = sns.barplot(data = df_sales_year, y='vendas', x = df_sales_year.index, palette = color)

  ax.set_title("Department stores' sales \n 2016 to 2049", loc="left", fontsize = 18, color = GREY1) # define title
  ax.set_xlabel(" ") # define x label
  ax.set_ylabel(" ") # define y label
  ax.xaxis.set_tick_params(labelsize=14, labelcolor=GRAY2) # define x axis tick size and color
  sns.despine(left=True, bottom = True) # remove spines

  ax.set_yticklabels([]) # remove y tick label
  for i, valor in enumerate(df_sales_year["vendas"]):
    qtd = f'R$ {valor:,.0f}'.replace(",",".")
    offset = 1e5
    ax.text(i, valor + offset, qtd, color = GREY2, fontsize = 12, ha = "center", va = "center") # create data labels

  return ax

ax = sales_visualization()
```

```python
colors = []
for year in df_sales_year.index:
  if df_sales_year.loc[year, "vendas"] == df_sales_year["vendas"].max():
    colors.append(BLUE2)
  else:
    colors.append(BLUE4)

ax = sales_visualization(colors) #define focus in the higgest value (2019)

ax.text(3.5, 1.5e6,
         'In $\\bf{2019}$, store\n'
         'sales increased\n'
         'arround $\\bf{22,3}$%\n'
         'compared to 2018.',
         fontsize=14, linespacing=1.45, color=BLUE2) # define comment

fig = ax.get_figure()
```

# Question 2: What are the top 7 most profitable products in our catalogue during the period shown?


```python
top_profit = sales.copy()
top_profit = top_profit[["tipo_produto", "lucro"]] # filter columns of interest

top_profit = top_profit.groupby("tipo_produto").sum() #grouping values
top_profit = top_profit.sort_values(by="lucro", ascending=False) # put in ascending format
top_profit = top_profit.head(7)

top_7 = top_profit[:7] # preparing the table that will generate the graph
top_7
```

```python
def top_7_visualization():

  # Área do gráfico e tema da visualização
  fig, ax = plt.subplots(figsize=(10,4))
  fig.subplots_adjust(right=0.7)
  sns.set_theme(style="white")

  colors = [BLUE2, BLUE2, BLUE2, GREEN1, GREEN1, GREY3, GREEN1]

  ax = sns.barplot(data = top_7, x = "lucro", y = top_7.index, palette = colors)

  ax.set_title('Top 7 most profitable products in the catalogue \n(2016-2019)\n', fontsize=18, color=GREY1, loc='left')
  ax.set_xlabel('')
  ax.set_ylabel('')
  ax.set_xticklabels([])
  ax.yaxis.set_tick_params(labelsize=10, labelcolor = GRAY2 )
  sns.despine(left = True, bottom = True)

  for i, valor in enumerate(top_7['lucro']):
      qtd = f'R$ {valor:,.0f}'.replace(',','.')
      offset = 1e3  # offset de 1.000
      ax.text(valor - offset, i, qtd, color= GREY5, fontsize=10, fontweight='bold', ha='right', va='center')

  return ax

ax = top_7_visualization()
```

Conclusion:

There were 3 departments that made the top 7 most profitable, with Automotive items leading the way (blue), followed by Gardening and Landscaping (green) and just one product from General Items (grey) which is "Ferramentas"
